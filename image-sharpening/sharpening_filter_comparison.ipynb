```python
##canny edge
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('/content/sample_data/full-moon.jpg', 0)

# Apply Canny
edges = cv2.Canny(img, 100, 200)

# Display
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```

```python
# Experiment 5: Highboost Filtering

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('/content/sample_data/full-moon.jpg', 0)

# 1. Define the 'blur' variable (required for the formula)
blur = cv2.GaussianBlur(img, (3,3), 0)

# 2. Set the boosting factor
A = 2.5 # Try different values like 1.0, 1.5, 3.0

# 3. Apply the Highboost formula
# Corrected typo: 'addweighted' -> 'addWeighted'
highboost = cv2.addWeighted(img, A, blur, -1, 0)

# Display
plt.imshow(highboost, cmap='gray')
plt.title("Highboost Filtering (A=2.5)")
plt.axis('off')
plt.show()
```
