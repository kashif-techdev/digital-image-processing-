```python
## laplacian sharpening

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('/content/laplacian_edge_detection.jpg', 0)

# Laplacian
lap = cv2.Laplacian(img, cv2.CV_64F)

# Sharpen
sharp = img - lap

# Convert to uint8
sharp = np.clip(sharp, 0, 255)
sharp = np.uint8(sharp)

# Display
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(sharp, cmap='gray')
plt.title("Sharpened")

plt.show()
```
