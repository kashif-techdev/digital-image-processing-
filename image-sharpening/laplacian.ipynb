```
# laplacian

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (0 reads it in grayscale)
img = cv2.imread('/image-sharpening/laplacian_edge_detection.jpg', 0)

# Apply Laplacian
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Display
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian Edge Detection")
plt.axis('off')
plt.show()

```
