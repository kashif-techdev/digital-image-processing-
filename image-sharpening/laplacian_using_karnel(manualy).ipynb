```python
# Experiment: 8 Neighbors Laplacian

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('/content/laplacian_edge_detection.jpg', 0)

# Define 8-neighbour Laplacian kernel
kernel_8 = np.array([[-1, -1, -1],
                     [-1,  8, -1],
                     [-1, -1, -1]])

# Apply filter
laplacian_8 = cv2.filter2D(img, -1, kernel_8)

# Display
plt.imshow(laplacian_8, cmap='gray')
plt.title("8-Neighbour Laplacian")
plt.axis('off')
plt.show()
```
